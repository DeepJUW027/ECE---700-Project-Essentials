```python
import os
import matplotlib.pyplot as plt
import cv2

BASE_DIR = "/home/deep-jariwala/Desktop/ECE_700_Fall_2025/Medical_Image_processing/RSNA_CT_SCAN_Detection/remote_desktop_execution/Interpretation_Results"

# Valid expected image files
expected_images = [
    "grey_input.png",
    "grey_gradcam_heatmap.png",
    "grey_gradcam_overlay.png",
    "grey_saliency.png",
    "pseudo_input.png",
    "pseudo_gradcam_heatmap.png",
    "pseudo_gradcam_overlay.png",
    "pseudo_saliency.png",
]

# Function to load image safely
def load_img(path):
    if not os.path.exists(path):
        return None
    img = cv2.imread(path)
    if img is None:
        return None
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    return img

# Function to display a full panel for one filename
def display_case(case_folder):
    case_path = os.path.join(BASE_DIR, case_folder)
    if not os.path.isdir(case_path):
        print(f"‚ùå Folder not found: {case_path}")
        return

    print(f"üìÇ Viewing: {case_folder}")

    fig, axes = plt.subplots(2, 4, figsize=(22, 10))
    axes = axes.flatten()

    for idx, img_name in enumerate(expected_images):
        img_path = os.path.join(case_path, img_name)
        img = load_img(img_path)

        if img is None:
            axes[idx].text(0.5, 0.5, f"Missing: {img_name}", fontsize=12, ha='center')
            axes[idx].axis("off")
        else:
            axes[idx].imshow(img)
            axes[idx].set_title(img_name, fontsize=10)
            axes[idx].axis("off")

    plt.tight_layout()
    plt.show()

# List available folders (each folder = one filename case)
cases = sorted([d for d in os.listdir(BASE_DIR) if os.path.isdir(os.path.join(BASE_DIR, d))])

print(f"Found {len(cases)} cases")
for i, c in enumerate(cases[:10]):
    print(f"{i}: {c}")

print("\nRun: display_case(cases[index]) to visualize a case.")
```
